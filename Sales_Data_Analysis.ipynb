```python
# ==============================
# Sales Data Analysis - Step 1
# ==============================

# Import Library
import pandas as pd
import os

# Create Sample Dataset
sales_data = {
    "Order_ID": [1001, 1002, 1003, 1004, 1005],
    "Date": ["2026-01-01", "2026-01-02", "2026-01-03", "2026-01-04", "2026-01-05"],
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Laptop"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics"],
    "Quantity": [2, 5, 3, 1, 1],
    "Price": [50000, 600, 1500, 12000, 50000],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", "Bangalore"]
}

# Convert Dictionary to DataFrame
df = pd.DataFrame(sales_data)

# Display Dataset
print("Sample Dataset:")
display(df)

# Save CSV in Current Folder
df.to_csv("sales_data.csv", index=False)

print("\n✅ Dataset saved successfully as 'sales_data.csv'")

# Read CSV Again
df = pd.read_csv("sales_data.csv")

print("\nCSV Loaded Successfully:")
display(df.head())
```

```python
# ==============================
# Step 2 - Explore the Dataset
# ==============================

import pandas as pd

# Load Dataset
df = pd.read_csv("sales_data.csv")

# First 5 Rows
print("First 5 Rows")
display(df.head())

# Last 5 Rows
print("Last 5 Rows")
display(df.tail())

# Dataset Shape
print("Shape of Dataset:")
print(df.shape)

# Column Names
print("\nColumn Names:")
print(df.columns)

# Data Types
print("\nData Types:")
print(df.dtypes)

# Dataset Information
print("\nDataset Information:")
df.info()

# Statistical Summary
print("\nStatistical Summary:")
display(df.describe())
```

```python
# ==============================
# Step 3 - Sales Analysis
# ==============================

import pandas as pd

# Load Dataset
df = pd.read_csv("sales_data.csv")

# Create Total Column
df["Total"] = df["Quantity"] * df["Price"]

print("Dataset with Total Column")
display(df)

# Overall Sales
total_sales = df["Total"].sum()
print("\nTotal Sales:", total_sales)

# Product-wise Sales
print("\nProduct-wise Sales")
product_sales = df.groupby("Product")["Total"].sum()
display(product_sales)

# City-wise Sales
print("\nCity-wise Sales")
city_sales = df.groupby("City")["Total"].sum()
display(city_sales)

# Category-wise Sales
print("\nCategory-wise Sales")
category_sales = df.groupby("Category")["Total"].sum()
display(category_sales)
```

```python
# ==============================
# Step 4 - Data Visualization
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("sales_data.csv")

# Create Total Column
df["Total"] = df["Quantity"] * df["Price"]

# ------------------------------
# Chart 1 - Product Wise Sales
# ------------------------------

product_sales = df.groupby("Product")["Total"].sum()

plt.figure(figsize=(8,5))
plt.bar(product_sales.index, product_sales.values)

plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")

plt.show()

# ------------------------------
# Chart 2 - City Wise Sales
# ------------------------------

city_sales = df.groupby("City")["Total"].sum()

plt.figure(figsize=(8,5))
plt.bar(city_sales.index, city_sales.values)

plt.title("City Wise Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")

plt.show()

# ------------------------------
# Chart 3 - Product Sales Pie Chart
# ------------------------------

plt.figure(figsize=(6,6))

plt.pie(product_sales.values,
        labels=product_sales.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Product Sales Percentage")

plt.show()
```

```python
# ==============================
# Step 5 - Dashboard Report
# ==============================

import pandas as pd

# Load Dataset
df = pd.read_csv("sales_data.csv")

# Create Total Column
df["Total"] = df["Quantity"] * df["Price"]

# -----------------------------
# Dashboard KPIs
# -----------------------------

total_sales = df["Total"].sum()
total_orders = len(df)
total_products = df["Product"].nunique()
total_cities = df["City"].nunique()
average_sales = df["Total"].mean()

print("=" * 50)
print("       SALES DASHBOARD")
print("=" * 50)

print(f"Total Sales      : ₹{total_sales:,.2f}")
print(f"Total Orders     : {total_orders}")
print(f"Total Products   : {total_products}")
print(f"Total Cities     : {total_cities}")
print(f"Average Sale     : ₹{average_sales:,.2f}")

print("=" * 50)

# -----------------------------
# Top Selling Products
# -----------------------------

print("\nTop Selling Products")

top_products = df.groupby("Product")["Total"].sum().sort_values(ascending=False)

print(top_products)

# -----------------------------
# City Wise Sales
# -----------------------------

print("\nCity Wise Sales")

city_sales = df.groupby("City")["Total"].sum().sort_values(ascending=False)

print(city_sales)

print("\nDashboard Generated Successfully!")
```

```python
# ==============================
# Step 6 - Save Charts
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import os

# Create charts folder
os.makedirs("charts", exist_ok=True)

# Load Dataset
df = pd.read_csv("sales_data.csv")

# Create Total Column
df["Total"] = df["Quantity"] * df["Price"]

# Product Wise Sales
product_sales = df.groupby("Product")["Total"].sum()

plt.figure(figsize=(8,5))
plt.bar(product_sales.index, product_sales.values)
plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.savefig("charts/product_sales.png")
plt.show()

# City Wise Sales
city_sales = df.groupby("City")["Total"].sum()

plt.figure(figsize=(8,5))
plt.bar(city_sales.index, city_sales.values)
plt.title("City Wise Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.savefig("charts/city_sales.png")
plt.show()

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(product_sales.values,
        labels=product_sales.index,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Product Sales Percentage")
plt.savefig("charts/product_sales_pie.png")
plt.show()

print("✅ Charts saved successfully in the 'charts' folder.")
```

```python
# ==============================
# Step 7 - Export Sales Report
# ==============================

import pandas as pd
import os

# Create output folder
os.makedirs("output", exist_ok=True)

# Load Dataset
df = pd.read_csv("sales_data.csv")

# Create Total Column
df["Total"] = df["Quantity"] * df["Price"]

# Product Wise Sales
product_sales = df.groupby("Product")["Total"].sum().reset_index()

# City Wise Sales
city_sales = df.groupby("City")["Total"].sum().reset_index()

# Save Reports
product_sales.to_csv("output/product_sales_report.csv", index=False)
city_sales.to_csv("output/city_sales_report.csv", index=False)

print("✅ Reports saved successfully!")

print("\nProduct Sales Report")
display(product_sales)

print("\nCity Sales Report")
display(city_sales)
```
